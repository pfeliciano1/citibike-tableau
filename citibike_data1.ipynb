```python
# Dependencies
import pandas as pd
```

```python
# Name of the CSV file
file = 'Resources/JC-202101-citibike-tripdata.csv'
```

```python
# Taking the file and converting into dataframe
df = pd.read_csv(file, encoding="ISO-8859-1")
df
```

```python
# Convert tripduration column from seconds to minutes
# df["tripduration"] = (df["tripduration"])/60
df['tripduration'] = pd.to_datetime(df["tripduration"], unit='s').dt.strftime("%H:%M:%S")
df
```

```python
# Replace gender values: 0 = unknown, 1 = male, 2 = female
df['gender'] = df['gender'].map({0:'unknown', 1:'male', 2:'female'})
df
```

```python
df.dtypes
```

```python
df['start date'] = pd.to_datetime(df['starttime']).dt.date
df['start time'] = pd.to_datetime(df['starttime']).dt.time
df['end date'] = pd.to_datetime(df['stoptime']).dt.date
df['end time'] = pd.to_datetime(df['stoptime']).dt.time
df
```

```python
# Identify incomplete rows
df.count()
```

```python
df = df.rename(columns={"tripduration":"trip duration", "usertype": "user type",
                        "starttime":"start time", "stoptime": "stop time"})
df
```

```python
# Export final wildfires_data file as a CSV
df.to_csv("Resources/citibike-data-202101.csv", index=False, header=True)
```
