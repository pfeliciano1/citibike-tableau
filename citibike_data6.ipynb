```python
# Dependencies
import pandas as pd
```

```python
# Name of the CSV file
file = 'Resources/JC-202106-citibike-tripdata.csv'
```

```python
# Taking the file and converting into dataframe
df = pd.read_csv(file, encoding="ISO-8859-1")
df
```

```python
df.dtypes
```

```python
# Identify incomplete rows
df.count()
```

```python
# Drop all rows with missing information
clean_df = df.dropna(how='any')
clean_df
```

```python
# Identify incomplete rows
clean_df.count()
```

```python
clean_df['start date'] = pd.to_datetime(clean_df['started_at']).dt.date
clean_df['start time'] = pd.to_datetime(clean_df['started_at']).dt.time
clean_df['end date'] = pd.to_datetime(clean_df['ended_at']).dt.date
clean_df['end time'] = pd.to_datetime(clean_df['ended_at']).dt.time
clean_df
```

```python
renamed_df = clean_df.rename(columns={"ride_id": "rideID", "rideable_type": "rideable type", 
                                      "started_at": "start timestamp", "ended_at": "end timestamp",
                                      "start_station_name": "start station name", 
                                      "start_station_id": "start station ID", "end_station_name": "end station name", 
                                      "end_station_id": "end station ID", "start_lat": "start latitude",
                                      "start_lng": "start longitude", "end_lat": "end latitude", 
                                      "end_lng": "end longitude", "member_casual": " member/casual"})
renamed_df
```

```python
renamed_df["rideable type"].value_counts()
```

```python
# Change name of value "docked_bike" in "rideable type" column
renamed_df["rideable type"] = renamed_df["rideable type"].replace(["docked_bike"],["docked bike"])
renamed_df["rideable type"] = renamed_df["rideable type"].replace(["classic_bike"],["classic bike"])
renamed_df
```

```python
# Export final wildfires_data file as a CSV
renamed_df.to_csv("Resources/citibike-data-202106.csv", index=False, header=True)
```
